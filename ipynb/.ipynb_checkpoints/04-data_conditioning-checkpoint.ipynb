### Boston sample (n=5) statistics

```r
boston_samp <- sample_n(boston, 5)
```

```r
features <- c('CRIM', 'INDUS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT')

boston_samp_stats <- data.frame(features, describe(boston_samp))
boston_samp_stats <- subset(boston_samp_stats, select=c('features','mean', 'sd', 'skew', 'kurtosis'))
boston_samp_stats
```

### Boston dataframe statistics

```r
boston_stats <- data.frame(features, describe(boston))
boston_stats <- subset(boston_stats, select=c('features','mean', 'sd', 'skew', 'kurtosis'))
boston_stats
```

### Scaling Data

To calculate the Z-score of our sample mean, we will use the following equation: 

$$Z = \frac{X-\mu}{\sigma}$$

where, 
* X = sample mean
* $\mu$ = population mean
* $\sigma$ = population sd

```r
boston_scaled <- data.frame(scale(boston))
```

```r
boston_scaled_stats <- data.frame(features, describe(boston_scaled))
boston_scaled_stats <- subset(boston_scaled_stats, select=c('features','mean', 'sd', 'skew', 'kurtosis'))
boston_scaled_stats
```

After scaling the data, the mean is approx 0, and the standard deviation is 1. Skew remains identical to skew in the original data.

```r
library(ggplot2)
library(repr)
options(repr.plot.width=10, repr.plot.height=2) 
```

```r
source('multiplot.r')
```

```r
original_v_scaled <- function (features) {
    original_feature <- as.vector(boston[[features]])
    scaled_feature <- as.vector(boston_scaled[[features]])
    p1 <- qplot(original_feature, geom="histogram", bins=200, alpha=I(.4),)+
        geom_vline(aes(xintercept=mean(original_feature)), color="red", linetype="dashed", size=1)+
        geom_vline(aes(xintercept=median(original_feature)), color="blue", linetype="dashed", size=1)+
        ggtitle(paste("Original Distribution of", features))
    p2 <- qplot(scaled_feature, geom="histogram", bins=200, alpha=I(.4),)+
        geom_vline(aes(xintercept=mean(scaled_feature)), color="red", linetype="dashed", size=1)+
        geom_vline(aes(xintercept=median(scaled_feature)), color="blue", linetype="dashed", size=1)  +
        ggtitle(paste("Scaled Distribution of", features))  
    multiplot(p1, p2, cols=2)
}
```

```r
original_v_scaled('CRIM')
original_v_scaled('INDUS')
original_v_scaled('NOX')
original_v_scaled('RM')
original_v_scaled('AGE')
original_v_scaled('DIS')
original_v_scaled('RAD')
original_v_scaled('TAX')
original_v_scaled('PTRATIO')
original_v_scaled('B')
original_v_scaled('LSTAT')
```

### Deskewing Data

We will use the log transform.

```r
boston_log <- log(boston)
```

```r
boston_log_scaled <- data.frame(scale(boston_log))
```

```r
skewness_ <- data.frame(features=colnames(boston_scaled))
skewness_['skew_log'] <- sapply(boston_log, skewness)
skewness_['skew_log_sc'] <- sapply(boston_log_scaled, skewness)
```
